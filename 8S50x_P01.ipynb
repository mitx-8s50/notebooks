<h3>Random Variables</h3>

We will begin with the concept of random variable, which is essential to probability and statistics.

<u>Definition:</u> A <b>random variable</b> is (for the purposes of this class) an observable whose outcome may differ between experiments.

For example, if you're measuring the length of a cube in your experiment, the length you measure is a random variable. This is because your measured results conceivably might vary, depending on how you take the measurement and the precision of the ruler, or even thermal expansion and contraction of the cube.

If you're measuring the number of snakes in a population, the number of snakes you count is also a random variable.

There are two types of random variables, continuous and discrete. Discrete random variables can only take on discrete values (snake example), whereas continuous random variables can take on continuous values (cube example).

<i>Technical note:</i> A random variable represents any possible values that the measurement could produce. The actual measured results are not random variables; they are data points. We often represent random variables by capital letters and their measured values as lowercase letters.

<br>
<!--end-block-->

<!--start-block-->
 
<h3>Probability Density Functions</h3>
 
Discrete random variables are associated with a probability mass function that defines the exact probability that the random variable will take a certain discrete value. More precisely,

<u>Definition:</u> The <b>probability mass function (PMF)</b> for a discrete random variable $R$ is a function $p(r)$ that maps from possible values $r$ of $R$ to the probability that $R=r$.</p>

<u>Corollaries:</u>
- The sum of $p(r)$ over all possible values of $R$ is 1. 
- $p(r)$ is unitless
- $p(r)$ is nonnegative
- $p(r) \leq 1$

For continuous probability distributions, we have instead a probability density function.

<u>Definition:</u>The <b>probability density funtion (PDF)</b> for a continuous random variable $R$ is a function $P(r)$ defined on possible values $r$ of $R$, such that the probability that $R$ is between $a$ and $b$ is given by</p>

$$p(a < R < b) = \int_{a}^{b} P(r) dr.$$

<u>Corollaries:</u>
- The integral of $P(r)$ over all possible values of $R$ is 1. 
- $P(r)$ has units of $1/r$
- $P(r)$ is nonnegative

Here, we adopt the convention that lowercase $p$ represent probabilities, and uppercase $P$ denote probability densities.

## <h2 style="border:1px; border-style:solid; padding: 0.25em; color: #FFFFFF; background-color: #90409C">P1.3 Joint Probability Density Functions</h2>

| [Top](#section_1_0) | [Previous Section](#section_1_2) | [Problems](#exercises_1_3) | [Next Section](#section_1_4) |

<h3>Overview</h3>

We can define joint probability density functions, e.g. $P(x,y)$, for multivariate probability distributions in a similar way. 

<u>Definition:</u> The <b>joint PDF</b> for two continuous random variables $X$ and $Y$ is a function $P(x, y)$ defined on possible values $x$ of $X$ and $y$ of $Y$, such that the probability that $X$ and $Y$ are in some region $R$ is given by</p>

$$p((X, Y)\in R) = \iint_{R} P(x, y) dx dy$$

We define the single variable PDFs $P(x)$ and $P(y)$ by integrating (or marginalizing) the extra variable from the joint PDF, i.e. 

$$P(x) = \int_{-\infty}^{\infty} P(x,y) dy$$ and $$P(y) = \int_{-\infty}^{\infty} P(x,y) dx$$


(Here, we implicitly assumed that the possible values of $X$ and $Y$ are the whole real number line.)

Now we can introduce another very useful term in statistics: independence. Conceptually, two random variables are independent if measuring one tells you nothing about measuring the other. Mathematically, this is:

<u>Definition:</u> Two random varibales $X$ and $Y$ are <b>independent</b> from one another if the joint PDF factors into the product of the two single-variable PDFs:</i>

$$P(x, y) = P(x) P(y).$$

<!--start-block-->

Given $k$ independent Gaussian random variables, the above definition of independence guarantees that the joint PDF is


$$P(x_1, x_2, ... x_k) = \prod_{i=1}^k P(x_i) = \frac{1}{ (2\pi)^{k/2}\prod_i \sigma_i} \exp \left(-\frac{1}{2}\sum_i \frac{(x_i- \mu_i)^2}{\sigma_i^2}\right)$$

<h3>Overview</h3>

Another very common distribution in physics is the Poisson distribution.

<u>Definition:</u> For a sequence of events that are independent of each other and independent of time, a random variable representing the number of events occuring within a given interval obeys the <b>Poisson distribution</b>.</p>

Examples include 
- the number of radioactive decays within a material over one second 
- the number of photons detected from a distant star in one day
- the number of meteorite impacts on Earth in one year

Note that Poisson-distributed random variables are discrete.

The PMF of a Poisson-distributed random variable with unitless parameter $\lambda$ is

$$p(k)=\frac{\lambda^k e^{-\lambda}}{k!}.$$

You can view $\lambda$ as the average rate of events times the time interval over which $k$ events are detected:

$$\lambda = \mathrm{(events\,per\,time)}\times\mathrm{(time\,interval)}$$

The variable $k$ is the number of events in the time period.

<br>





A Poisson distribution approaches a Gaussian distribution for large values of $\lambda$.

<br>

<h3>Example</h3>

Run the following cell to see several examples of Poisson distributions with different values of $\lambda$. Each histogram is generated from 1500 samples of a Poisson distribution. A continuous Gaussian distribution and Gaussian mean are also plotted in each figure.

Change the values of $\lambda$ in the list at your discretion.

```python
#>>>RUN

def plot_many_Poisson(l):
    hists = {} # Dictionary that will contain the Poisson-distributed data, listed by l value

    #plotting-------------------
    fig, ax = plt.subplots(nrows=4, ncols=1, figsize=(8,30))
    for k in range(len(ax)):
        #plot data
        s = np.random.poisson(l[k],1500)
        hists[l[k]]=s
        bins = np.arange(0, np.max(s) + 1, 1) - 0.5
        ax[k].hist(s, bins=bins, density=True, label="Poisson data")
        x = np.linspace(min(s),max(s),100)
        ax[k].plot(x,(1/(np.sqrt(2*np.pi*l[k])))*np.exp(-0.5*(x-l[k])**2.0/l[k]), label="Gaussian dist.")
        ax[k].axvline(x=l[k],c='r', label="Gaussian mean")
        
        #plot2 labels and style
        ax[k].legend(loc="upper right", fontsize=12)
        ax[k].set_xlabel("Counts in interval",fontsize=15)
        ax[k].set_ylabel("Probability of observation",fontsize=15)
        ax[k].set_title(f"lambda={l[k]}", fontsize=15)
        
        plt.setp(ax[k].get_xticklabels(), fontsize=12)
        plt.setp(ax[k].get_yticklabels(), fontsize=12)
        
        # plot grid
        ax[k].grid(alpha=0.2)


        
# l is a list of lambda values
l = [1,5,10,25] 
plot_many_Poisson(l)
```

<h3>Characteristics of Poisson Processes</h3>

Let's simulate a **Poisson process**, which is just a process leading to a set of events that have the following characteristics:
- they are independent of time, meaning that they are not more or less frequent depending on the time of detection
- they are independent of each other, meaning the probability distribution of one event is unaffected by the presence of other events.
- they do not occur at the same time.

A random variable that counts these events occuring within an interval is Poisson-distributed.

<br>

<h3>Choosing Random Event Times</h3>

## <h2 style="border:1px; border-style:solid; padding: 0.25em; color: #FFFFFF; background-color: #90409C">P1.6 Optional Exercise: Exponential distributions</h2>

<h3>Overview</h3>

For Poisson processes, we've seen that
* The probability distribution of each event's time of occurence is <b>uniform</b>
* The probability distribution of the number of events occuring within an interval is <b>Poisson</b>

However, we have not discussed the probability distribution of the time between events. It turns out to be exponential, with PDF

$$P(\Delta t)= r e^{-r \Delta t}$$

where $r$ is the average event rate.

<br>
<!--end-block-->

<h3>Example</h3>

Let's demonstrate this from our above data.

```python
#>>>RUN

def plot_new(NUM_EVENTS,TIME_RANGE,TIME_INTERVAL):
    event_times = random_event_data(NUM_EVENTS, TIME_RANGE)
    event_times.sort()

    bins = np.linspace(0, 1, 10)
    deltas = [event_times[i+1] - event_times[i] for i in range(NUM_EVENTS - 1)]
  
    rate = (NUM_EVENTS / TIME_RANGE)

    #plotting-------------------
    #plot data  
    plt.hist(deltas, density=True, label="Observed distribution", bins=bins)
    plt.plot(bins, np.exp(-rate * bins) * rate)
    
    #plot labels and style
    plt.title('', fontsize = 15)
    plt.xlabel("Time between events", fontsize = 15)
    plt.ylabel("Fraction of events", fontsize = 15)
    
    # changing the fontsize of ticks
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    

NUM_EVENTS = 500
TIME_RANGE = 250
TIME_INTERVAL = 1
plot_new(NUM_EVENTS,TIME_RANGE,TIME_INTERVAL)
```
